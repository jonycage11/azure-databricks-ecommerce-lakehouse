```python
RAW_PATH = "/Volumes/ecommerce_project/default/raw_data/"

files = {
    "customers": "olist_customers_dataset.csv",
    "orders": "olist_orders_dataset.csv",
    "order_items": "olist_order_items_dataset.csv",
    "products": "olist_products_dataset.csv",
    "payments": "olist_order_payments_dataset.csv",
}

for table_name, file_name in files.items():
    df = (
        spark.read
        .option("header", True)
        .option("inferSchema", True)
        .csv(RAW_PATH + file_name)
    )

    df.write.format("delta").mode("overwrite").saveAsTable(f"bronze_{table_name}")

    print(f"bronze_{table_name}: {df.count()} rows loaded")

for table_name in files.keys():
    print(f"--- bronze_{table_name} ---")
    spark.sql(f"SELECT * FROM bronze_{table_name} LIMIT 3").show()
```
